```python
import pandas as pd 
import numpy as np
```

```python
data = {
    'Product': ['Bag', 'Pen', 'Notebook', 'Bag', 'Pen',
                'Notebook', 'Bag', 'Pen', 'Notebook', 'Bag',
                'Pen', 'Notebook', 'Bag', 'Pen', 'Notebook',
                'Bag', 'Pen', 'Notebook', 'Bag', 'Pen'],
    'Quantity': [5, 10, 7, 3, 8,
                 6, 4, 9, 12, 2,
                 5, 11, 8, 3, 6,
                 9, 4, 7, 10, 5],
    'Price': [20, 5, 8, 22, 5,
              8, 25, 6, 10, 18,
              22, 6, 9, 20, 5,
              11, 27, 15, 19, 8]
}
```

```python
#Product => products of sales
#Quantity => How many take products?
#Price => product's price
#total => Quantity*Price
```

```python
df=pd.DataFrame(data=data)
df["total"]=df["Quantity"]*df["Price"]
df
```

```python
df.info()
```

```python
df["Product"].unique()
```

```python
#using groupby
#using pivot table
#using crosstab() 
df.head()
```

#### Using Groupby

```python
##Aggregating Sum and Mean 
df.groupby("Product").agg({"Price":["sum","mean"]})
```

```python
##Multiple Aggregations on Different Columns
df.groupby("Product").agg({"total":"sum","Quantity":"mean"})
```

```python
from statistics import stdev
df.groupby("Product").agg({"Price":["sum","mean",stdev]})
#This applied the standard deviaton function "stdev" from Python's statistics module
```

```python
df.head()
```

```python
##Aggregations After Transform
#Transform the data 
#Aggregate on the transformed data
df["Squared"]=df["Price"]**2
df.groupby("Product").agg({"Squared":"sum"})
#This aggregates the sum of the squared values per group,allowing for different metrics
```

```python
grouped=df.groupby("Product")
grouped
```

```python
##Applying Aggregations Functions

# Calculate the sum for each group
sum_values=grouped["total"].sum()
# Calculate the mean for each group
mean_values=grouped["total"].mean()
# Find the minimum and maximum values for each group
min_max_values=grouped["total"].agg(["min","max"])
# Calculate the size (number of elements) in each group
size_elements=grouped["total"].size()

sum_values,mean_values,min_max_values,size_elements
```

```python
#Grouping by Multiple Column
multi_grouped=df.groupby(["Product","Quantity"])
multi_grouped
```

```python
#Resetting the Index
###Why Reset Index?
#Enhanced Readability
#Facilating Further Analysis
df_reset=df.reset_index()
df_reset
```

```python
##Crosstab() Funciton => Compute a frequency table of two or more factors.
pd.crosstab(df["Product"],df["Quantity"])
```

```python
##Pivot_table() Function => used to Summarize and aggregate data in a DataFrame
pivot_table=pd.pivot_table(df,values="Quantity",index="Product")
pivot_table
```

```python
#Custom Aggregation Functions => to perform spesific calculations unique to your analysis.
def diff_func(x):
    return x.mean()/x.sum()
custom_agg=grouped["total"].agg(diff_func)
custom_agg
```
